```python
for team in all_team5:
    driver = webdriver.Chrome()
    driver.get('https://www.onlinegambling.com/nba-gameday-calculator/')
    driver.maximize_window()
    team_insert = driver.find_element_by_class_name("search__input")
    team_insert.send_keys(team)

    time.sleep(2)

    suggestion = driver.find_element_by_class_name('autocomplete__result')
    suggestion.click()


    enter = driver.find_element_by_class_name('nba-gc__footer-icon')
    enter.click()
    # two standard tickets
    standard_ticket = driver.find_element_by_xpath('//*[@id="nba-gameday-calculator-widget"]/div[2]/div[2]/div[2]/div[3]/button[1]')
    standard_ticket.click()
    standard_ticket.click()

    enter = driver.find_element_by_class_name('nba-gc__footer-icon')
    enter.click()
    # two beers , one nacho
    standard_ticket = driver.find_element_by_xpath('//*[@id="nba-gameday-calculator-widget"]/div[2]/div[2]/div[1]/div[3]/button[1]')
    standard_ticket.click()
    standard_ticket.click()

    standard_ticket = driver.find_element_by_xpath('//*[@id="nba-gameday-calculator-widget"]/div[2]/div[2]/div[4]/div[3]/button[1]')
    standard_ticket.click()

    enter = driver.find_element_by_class_name('nba-gc__footer-icon')
    enter.click()

    # one jersey one hat
    standard_ticket = driver.find_element_by_xpath('//*[@id="nba-gameday-calculator-widget"]/div[2]/div[2]/div[1]/div[3]/button[1]')
    standard_ticket.click()

    standard_ticket = driver.find_element_by_xpath('//*[@id="nba-gameday-calculator-widget"]/div[2]/div[2]/div[3]/div[3]/button[1]')
    standard_ticket.click()

    enter = driver.find_element_by_class_name('nba-gc__footer-icon')
    enter.click()

    # no parking
    driver.find_elements_by_class_name("step-calc")[2].click()
    enter = driver.find_element_by_class_name('nba-gc__footer-icon')
    enter.click()


    html = driver.page_source
    driver.close()


    soup = BeautifulSoup(html,'html.parser')



    team_name = soup.find(name='div',attrs={"class":"result__item"}).text

    all_price=[]
    for i in soup.find_all(name='div',attrs={"class":"result__price"}):
        all_price.append(float(i.text.strip()[1::]))

    team_ticket[team_name]=all_price[0]
    team_snack[team_name]=all_price[1]
    team_jersey[team_name]=all_price[2]
```
