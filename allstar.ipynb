```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('todo.csv')
```

```python
pd.options.display.max_rows = 50000
```

```python
data
```

```python
plt.style.use('fivethirtyeight')
```

```python
def draw(column,a,ht,color):
    fig = plt.gcf()
    fig.set_size_inches(18.5,10.5)

    data_draw = data.sort_values(column,ascending=False)

    rank = np.arange(5)

    plt.bar(rank,data_draw[column].to_list()[0:5],color = color,width = 0.35)

    plt.xticks(rank,data_draw['Player'].to_list()[0:5],fontsize=20,rotation = -15)
    
    plt.legend([column],prop={'size':20})

    a = a
    for i in data_draw[column].to_list()[0:5]:
        plt.text(a,i+ht,str(int(i)),fontsize=20)
        a+=1
```

```python
draw('PTS',-0.07,5,'#083D77')
```

```python
draw('TRB',-0.07,2,'#FF6978')
```

```python
draw('AST',-0.07,2,'#E9B44C')
```

```python
draw('BLK',-0.07,0.5,'#9B2915')
```

```python
draw('STL',-0.05,0.5,'#50A2A7')
```
