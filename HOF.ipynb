```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
pd.options.display.max_rows=50000
pd.options.display.max_columns=50000
```

```python
Kobe_url='https://www.basketball-reference.com/players/b/bryanko01.html'
Duncan_url='https://www.basketball-reference.com/players/d/duncati01.html'
Garnett_url='https://www.basketball-reference.com/players/g/garneke01.html'
```

# pts

```python
def get_pts_2p_3p(url):
    web = requests.get(url)

    content = web.content

    soup = BeautifulSoup(content,'html.parser')

    html = soup.find(name='div',attrs={"id":"div_per_game"})

    data = pd.read_html(str(html))[0]

    data = data[data['Season']=='Career']

    data = data.loc[:,['PTS','2P%','3P%']]

    data['2P%']=data['2P%']*100

    data['3P%']=data['3P%']*100

    return data.values.tolist()[0]
```

```python
Kobe_pts_data = get_pts_2p_3p(Kobe_url)

Duncan_pts_data = get_pts_2p_3p(Duncan_url)

Garnett_pts_data = get_pts_2p_3p(Garnett_url)
```

```python
plt.style.use('fivethirtyeight')
```

```python
N = 3
ind = np.arange(N)
width = 0.2

fig = plt.figure()
ax = fig.add_subplot(111)
fig.set_size_inches(18.5,10.5)

kobe=Kobe_pts_data
rect1 = ax.bar(ind,kobe,width,color='#FDB927')
duncan=Duncan_pts_data
rect2 = ax.bar(ind+width,duncan,width,color='#444444')
garnett=Garnett_pts_data
rect3 = ax.bar(ind+width*2,garnett,width,color='#00843D')

ax.set_xticks(ind+width)
ax.set_xticklabels(('PTS','2P%','3P%'))
ax.tick_params(labelsize=25)

a=-0.06
for c in kobe:
    plt.text(a,c+1,(str(c)),fontsize=25)
    a+=1
a=0.13
for c in duncan:
    plt.text(a,c+1,(str(c)),fontsize=25)
    a+=1
a=0.33
for c in garnett:
    c=round(c,1)
#     ↑取小數點後一位
    plt.text(a,c+1,(str(c)),fontsize=25)
    a+=1
    
    plt.legend(['Kobe Bryant','Tim Duncan','Kevin Garnett'],loc = 0, prop = {'size':25})
```
